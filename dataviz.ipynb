```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# cores e fontes
plt.rcParams['font.family'] = 'Arial'

VERMELHO = '#C00C0C'
BEGE = '#CAAFAF'
CINZA = '#8C7F7F'
```

```python
# dados ficticios
X = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
novos_videos = [10, 13, 11, 15, 14, 20, 19, 15, 20, 18, 23, 16]
novos_membros = [10, 11, 8, 9, 7, 15, 17, 13, 18, 16, 20, 13]
```

## Primeiro plot

Primeiramente vamos manter o padrão que o Matplotlib apresenta.

```python
fig, ax = plt.subplots(figsize=(9,6), dpi=72)

ax.plot(range(12), novos_videos)
ax.plot(range(12), novos_membros)
ax.set_title('Novos membros do canal comparado ao número de novos vídeos mensais')

plt.show()
```

## Plot do gráfico ajustado

Com algumas mudanças diretamente no código, conseguimos um resultado muito melhor, sendo que o gráfico fica mais atraente e mais claro para interpretar.

```python
# plot da figura do gráfico
fig, ax = plt.subplots(figsize=(10,6), dpi=72)

# plot das linhas do gráfico
ax.plot(range(12), novos_videos, color = BEGE, linewidth = 4)
ax.plot(range(12), novos_membros, color = VERMELHO, linewidth = 4)

# eliminando as linhas de cima e da direita da figura
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# ajuste dos ticks do eixo x
ax.set_xticks(np.arange(12))
ax.set_xticklabels(X)

# eliminando os ticks do eixo y
ax.set_yticklabels([])
ax.set_yticks(ticks=[])

# ajuste de limite do eixo y
ax.set_ylim(0, 30)

# plot dos scatters selecionados em ambas as linhas
ax.scatter(range(12)[-3:], novos_videos[-3:], s = 90, color = BEGE, clip_on = False)
ax.scatter(range(12)[-3:], novos_membros[-3:], s = 90, color = VERMELHO, clip_on = False)

# adicionando os valores dos scatters
for i, j in enumerate(novos_videos[-3:]):
  ax.annotate(
      j,
      xy = (i + 9, j),
      xytext = (-7, 15),
      textcoords = 'offset points',
      color=BEGE,
      fontsize=12,
      fontweight='medium'
  )

for i, j in enumerate(novos_membros[-3:]):
  ax.annotate(
      j,
      xy = (i + 9, j),
      xytext = (-10, -22),
      textcoords = 'offset points',
      color=VERMELHO,
      fontsize=12,
      fontweight='medium'
  )


# legenda no final de cada linha
ax.annotate(
    "Novos Vídeos",
    xy=(len(novos_videos)-1, novos_videos[-1]),
    xytext=(40,-5),
    textcoords='offset points',
    color=BEGE,
    fontsize=18,
    fontweight='semibold'
)

ax.annotate(
    "Novos Membros",
    xy=(len(novos_membros)-1, novos_membros[-1]),
    xytext=(40,-5),
    textcoords='offset points',
    color=VERMELHO,
    fontsize=18,
    fontweight='semibold'
)

# título
ax.text(-0.5, 36,'Novos Membros do Canal', fontsize=24, color=VERMELHO, fontweight='bold')
ax.text(-0.5, 31,'Número de novos membros comparado\nao número de vídeos postados no\ncanal do YouTube a cada mês', fontsize=13, color=CINZA)

# rodapé
ax.text(
    -0.5,
    -5,
    "Fonte: Dados fictícios de um canal inexistente",
    fontsize=12, color=CINZA
)

# mostrando o gráfico
plt.show()
```

## Ainda podemos trabalhar muito mais em cima do design e dos detalhes do gráfico, mas com apenas algumas alterações simples já é muito melhor de "bater o olho" e conseguir entender sobre o que se trata, além de identificar de forma fácil e rápida as diferenças de valores.
